```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# read data, remove unused column, drop rows with missing values
data = pd.read_csv("data/disease.csv")
data
```

```python
symptoms_cols = data.iloc[:, 1:18]

symptoms = []

for col in symptoms_cols:
    for symptom in np.unique(symptoms_cols[col].astype("str")):
        if symptom not in symptoms:
            symptoms.append(symptom)
symptoms.sort()
symptoms.remove("nan")
```

```python
from math import isnan

ohe_data = {}

data_t = data.values.T
r, c = data_t.shape

ohe_data["disease"] = data_t[0]
sym_idx = 1

for symptom in symptoms:
    arr = np.empty(c, dtype=bool)
    for idx, sym in enumerate(data_t[sym_idx]):
        arr[idx] = (sym == symptom)
    ohe_data[symptom] = arr
    idx += 1

new_data = pd.DataFrame(data=ohe_data)
new_data
```

```python
def factorize_objs(df: pd.DataFrame):
    for colname, dtype in zip(df.keys(), df.dtypes.to_list()):
        if dtype == np.dtypes.ObjectDType:
            vals, keys = pd.factorize(df[colname])
            df[colname] = vals
            print(keys)

factorize_objs(new_data)
new_data
```

```python
target = "disease"
features = new_data.keys().to_list()
features.remove("disease")

X_train, X_test, y_train, y_test = train_test_split(new_data[features], new_data[target], train_size=0.8, random_state=42)
```

```python
from sklearn.naive_bayes import BernoulliNB

model = BernoulliNB(binarize=None, fit_prior=False, alpha=0.2)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
import dill
dill.settings["recurse"] = True

test = X_test.join(y_test)
test.to_csv("clean_data/disease_test.csv", index=False)

with open("models/disease_nb.modelfile", 'wb') as f:
    dill.dump(model, f)
```
